```python
import tkinter as tk

def open_new_window():
    new_window = tk.Toplevel(window)
    new_window.title("New Window")
    new_window.geometry("300x200")
    tk.Label(new_window, text="This is a new window!").pack(pady=10)

# Create the main window
window = tk.Tk()
window.title("Main Window")
window.geometry("500x300")

# Create a button in the main window to open the new window
open_button = tk.Button(window, text="Open New Window", command=open_new_window)
open_button.pack(pady=50)

# Start the application's main loop
window.mainloop()
```

```python
import customtkinter as ctk
from tkinter import messagebox
import math


class ScientificCalculator(ctk.CTk):
    def __init__(self):
        super().__init__()

        self.title("EA Scientific Calculator")
        self.geometry("500x720")
        self.resizable(False, False)

        # ---- Theme ----
        ctk.set_appearance_mode("dark")
        ctk.set_default_color_theme("blue")

        self.expression = ""
        self.is_degree_mode = True  # DEG mode default

        # ---- Display field ----
        self.text_input = ctk.StringVar()
        self.entry = ctk.CTkEntry(
            self,
            textvariable=self.text_input,
            font=("Consolas", 28, "bold"),
            justify="right",
            height=80,
            corner_radius=15,
            state="readonly"  # Prevent user from typing manually
        )
        self.entry.grid(row=0, column=0, columnspan=4, padx=15, pady=(20, 10), sticky="nsew")

        # ---- Mode toggle ----
        self.mode_button = ctk.CTkButton(
            self,
            text="Mode: DEG",
            corner_radius=20,
            fg_color="#f38ba8",
            hover_color="#f8cdda",
            text_color="black",
            command=self.toggle_mode
        )
        self.mode_button.grid(row=1, column=0, columnspan=4, padx=15, pady=5, sticky="nsew")

        # ---- Scientific Buttons ----
        sci_buttons = [
            ['sin', 'cos', 'tan', 'log'],
            ['ln', '√', '^', 'C'],
            ['π', 'e', '(', ')']
        ]

        # ---- Number/Operator Buttons ----
        num_buttons = [
            ['7', '8', '9', '/'],
            ['4', '5', '6', '*'],
            ['1', '2', '3', '-'],
            ['0', '.', '=', '+']
        ]

        # ---- Create Scientific Buttons ----
        row_index = 2
        for r in sci_buttons:
            col_index = 0
            for ch in r:
                self.create_button(ch, row_index, col_index, "#b4befe")
                col_index += 1
            row_index += 1

        # ---- Create Number Buttons ----
        for r in num_buttons:
            col_index = 0
            for ch in r:
                # Color scheme
                if ch == "=":
                    color = "#a6e3a1"  # green
                elif ch in ['/', '*', '-', '+']:
                    color = "#89b4fa"  # blue
                else:
                    color = "#313244"  # dark grey
                self.create_button(ch, row_index, col_index, color)
                col_index += 1
            row_index += 1

        # ---- Grid Configuration ----
        for i in range(row_index + 1):
            self.grid_rowconfigure(i, weight=1)
        for j in range(4):
            self.grid_columnconfigure(j, weight=1)

    # ---- Helper to Create Buttons ----
    def create_button(self, text, row, col, color):
        btn = ctk.CTkButton(
            self,
            text=text,
            corner_radius=18,
            fg_color=color,
            hover_color="#45475a",
            text_color="white",
            font=("Segoe UI", 16, "bold"),
            command=lambda: self.on_click(text)
        )
        btn.grid(row=row, column=col, padx=8, pady=8, sticky="nsew")

    # ---- Button Logic ----
    def on_click(self, char):
        if char == 'C':
            self.expression = ""
            self.text_input.set("")
        elif char == '=':
            self.evaluate()
        elif char == 'π':
            self.expression += str(math.pi)
            self.text_input.set(self.expression)
        elif char == 'e':
            self.expression += str(math.e)
            self.text_input.set(self.expression)
        elif char == '√':
            self.expression += "math.sqrt("
            self.text_input.set(self.expression)
        elif char == '^':
            self.expression += "**"
            self.text_input.set(self.expression)
        elif char in ['sin', 'cos', 'tan', 'log', 'ln']:
            func_map = {
                'sin': 'math.sin',
                'cos': 'math.cos',
                'tan': 'math.tan',
                'log': 'math.log10',
                'ln': 'math.log'
            }
            self.expression += f"{func_map[char]}("
            self.text_input.set(self.expression)
        else:
            self.expression += str(char)
            self.text_input.set(self.expression)

    # ---- Degree/Radian Toggle ----
    def toggle_mode(self):
        self.is_degree_mode = not self.is_degree_mode
        mode = "DEG" if self.is_degree_mode else "RAD"
        self.mode_button.configure(text=f"Mode: {mode}")

    # ---- Evaluate Expression ----
    def evaluate(self):
        try:
            expr = self.expression

            # Handle degree to radian conversion for trig functions if needed
            if self.is_degree_mode:
                # Wrap trig function arguments with math.radians
                # e.g. math.sin(x) -> math.sin(math.radians(x))
                expr = expr.replace("math.sin(", "math.sin(math.radians(")
                expr = expr.replace("math.cos(", "math.cos(math.radians(")
                expr = expr.replace("math.tan(", "math.tan(math.radians(")

            # Evaluate expression safely with only math module allowed
            result = eval(expr, {"math": math, "__builtins__": None})

            # Convert float result to int if no decimal part for cleaner display
            if isinstance(result, float) and result.is_integer():
                result = int(result)

            self.expression = str(result)
            self.text_input.set(self.expression)
        except Exception as e:
            messagebox.showerror("Error", f"Invalid Input:\n{str(e)}")
            self.expression = ""
            self.text_input.set("")


if __name__ == "__main__":
    app = ScientificCalculator()
    app.mainloop()
```
